### **TASACIÓN DE DEPARTAMENTOS**

**CONTEXTO GENERAL DEL CASO:**

**Empresa:** UrbanProp, una startup de tecnología inmobiliaria o "PropTech".

**Situación:** UrbanProp quiere revolucionar el mercado de compraventa de departamentos en la capital. Hasta ahora, las tasaciones se hacían de forma manual y subjetiva. La Gerencia Comercial necesita un algoritmo capaz de predecir el precio de venta de las propiedades basándose en sus características físicas y de ubicación.

El equipo de Data Science (ustedes) ha recibido una base de datos piloto `T3_B1_urbanprop.xlsx` con información de 468 departamentos. La base fue construida con fines académicos para analizar cómo distintas características físicas y de ubicación se relacionan con el precio de venta.

Su misión es construir y validar modelos estadísticos para explicar el precio.

El archivo contiene el siguiente conjunto de datos:
- `precio_millones`: Precio de venta en miles de pesos (Variable Dependiente $Y$).
- `superficie_m2`: Tamaño total del departamento en metros cuadrados.
- `calidad_terminaciones`: Puntaje de 1 a 10 evaluado por un inspector.
- `distancia_centro_km`: Distancia en kilómetros a un centro urbano de referencia.
- `habitaciones`: Número de dormitorios.
