#### (4) MODELO DE REGRESIÓN LINEAL MÚLTIPLE

(4.1) Definición de variables desde datos escritos

```python
# Variables independientes o predictoras
x1 = np.array([valor1, valor2, ...])
x2 = np.array([valor1, valor2, ...])
...
xk = np.array([valor1, valor2, ...])

# Crear un DataFrame con las variables predictoras
X = pd.DataFrame({
    'NOMBRE_X1': x1,
    'NOMBRE_X2': x2,
    ...
    'NOMBRE_XK': xk
})

# Variable dependiente o respuesta
y = np.array([valor1, valor2, ...])
```

(4.2) Definición de variables desde un dataset

```python
# Seleccionar la variable dependiente o respuesta
df["NOMBRE_VARIABLE_DEPENDIENTE"]

# Seleccionar las variables independientes o predictoras
X = df[["NOMBRE_VARIABLE_X1", "NOMBRE_VARIABLE_X2","...", "NOMBRE_VARIABLE_XK"]]
```

(4.3) Generación de OLS Regression Results

```python
# Agregar la constante al modelo
X = sm.add_constant(X)

# Ajustar el modelo de regresión lineal múltiple
modelo = sm.OLS(y, X).fit()

# Mostrar la tabla OLS
print(modelo.summary())
```

#### (5) ANÁLISIS DE RESIDUOS

(5.1) Cálculo de valores estimados y residuos

```python
# Calcular los valores estimados por el modelo
valores_estimados = modelo.predict(X)

# Calcular los residuos del modelo
residuos = y - valores_estimados
```

(5.2) Construcción de gráfico de residuos

```python
# Construir gráfico de residuos versus valores estimados
plt.scatter(valores_estimados, residuos)

# Agregar línea de referencia en cero
plt.axhline(0, linestyle='--')

# Personalización del gráfico
plt.xlabel('Valores estimados')
plt.ylabel('Residuos')
plt.title('Residuos vs valores estimados')
plt.grid()
plt.show()
```
